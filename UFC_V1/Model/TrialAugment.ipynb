```python
import pandas as pd

# Load the datasets
augmented_results_df = pd.read_csv('Augmented_Results.csv')
clean_data_df = pd.read_csv('CleanData2.csv')

```

```python
# Extract the names of the fighters from the Fight column in Augmented_Results
augmented_results_df['Fighter1'] = augmented_results_df['Fight'].apply(lambda x: x.split(' vs. ')[0].strip())
augmented_results_df['Fighter2'] = augmented_results_df['Fight'].apply(lambda x: x.split(' vs. ')[1].strip())

# Set of fighter names from CleanData2.csv
fighter_names = set(clean_data_df['Fighter Name'])

# Filter the Augmented_Results to include only rows where both fighters are in CleanData2.csv
filtered_augmented_results_df = augmented_results_df[
    (augmented_results_df['Fighter1'].isin(fighter_names)) & (augmented_results_df['Fighter2'].isin(fighter_names))
]

# Select relevant columns and display the filtered data
filtered_augmented_results_df = filtered_augmented_results_df[['Fight', 'OUTCOME']]
len(filtered_augmented_results_df)
```

```python
filtered_augmented_results_df
```

```python
# Retaining the Fighter1 and Fighter2 columns properly in the filtered_augmented_results_df
filtered_augmented_results_df['Fighter1'] = augmented_results_df['Fighter1']
filtered_augmented_results_df['Fighter2'] = augmented_results_df['Fighter2']

# Now displaying the first few entries correctly
final_fight_data_correct = filtered_augmented_results_df[['Fighter1', 'Fighter2', 'OUTCOME']]
final_fight_data_correct.head()
```

```python
from fuzzywuzzy import process

# Function to standardize names by removing spaces and converting to lowercase
def standardize_names(name):
    return ''.join(name.lower().split())

# Apply standardization to the fighter names in both datasets
clean_data_df['Standardized Name'] = clean_data_df['Fighter Name'].apply(standardize_names)
augmented_results_df['Standardized Fighter1'] = augmented_results_df['Fighter1'].apply(standardize_names)
augmented_results_df['Standardized Fighter2'] = augmented_results_df['Fighter2'].apply(standardize_names)

# Create a set of standardized names from CleanData2.csv for quick lookup
standardized_fighter_names = set(clean_data_df['Standardized Name'])

# Function to match fighters using the standardized names
def match_fighters(row):
    if row['Standardized Fighter1'] in standardized_fighter_names and row['Standardized Fighter2'] in standardized_fighter_names:
        return True
    return False

# Apply the matching function to filter the fights
filtered_fights = augmented_results_df[augmented_results_df.apply(match_fighters, axis=1)]
filtered_fights[['Fight', 'OUTCOME']]
```
