```python
import pandas as pd

# Load your DataFrame
df = pd.read_csv('Predictions.csv')

# Ensure all data in 'Real Outcome' and 'New Model Outcome' are in consistent case format
df['Real Outcome'] = df['Real Outcome'].str.lower()

# Initialize the 'Model Outcome' column
df['Model Outcome'] = None

for i in range(len(df)):
    try:
        real_outcome = df.at[i, 'Real Outcome']
        predicted_winner = df.at[i, 'New Model Outcome']
        fighter1 = df.at[i, 'Fighter 1']

        if real_outcome in ['w/l', 'l/w']:
            # Determine the model outcome based on the real outcome and predicted winner
            if (real_outcome == 'w/l' and predicted_winner == fighter1) or (real_outcome == 'l/w' and predicted_winner != fighter1):
                df.at[i, 'Model Outcome'] = 'Win'
            else:
                df.at[i, 'Model Outcome'] = 'Loss'
    except Exception as e:
        print(f"Error processing row {i}: {e}")

```

```python
df.to_csv("Predictions.csv")
```
