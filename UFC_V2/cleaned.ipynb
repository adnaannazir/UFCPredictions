```python
import pandas as pd
```

```python
df = pd.read_csv("NewCleanedDataDuplicated.csv")
```

```python
# Check for any other columns containing percentage signs
percentage_columns = [col for col in df.columns if df[col].dtype == 'object' and df[col].str.contains('%').any()]

# Convert percentage columns to float
for col in percentage_columns:
    df[col] = df[col].str.replace('%', '').astype(float) / 100

# Verify the conversion by displaying the updated data types and the first few rows
updated_types = df[percentage_columns].dtypes
df[percentage_columns].head(), updated_types
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.to_csv("FinalCleanedData.csv")
```
